## Gaussian Blur

```python
# Import necessary libraries
from PIL import Image
from PIL import ImageFilter
```

```python
# Load image
img = Image.open("images/image.jpg")
# Apply gaussian blur
blur_img = img.filter(ImageFilter.GaussianBlur(5))
# Save image
blur_img.save("Saved_images/GaussianBlur.jpg")
# Display image
blur_img.show()
```

# Gaussian Filter

```python
# Import necessary libraries
from skimage import io
from skimage import filters
```

```python
# Load image
img = io.imread("images/image.jpg")

# Apply filter 
out = filters.gaussian(img,sigma=5)

# Save and display image
io.imsave("Saved_images/gaussian filter scikit.jpg",out)
io.imshow(out)
io.show()
```

## Median Filter

```python
# Import necessary libraries
from PIL import Image
from PIL import ImageFilter
```

```python
# Load image
img = Image.open("images/input.png")
# Apply gaussian blur
blur_img = img.filter(ImageFilter.MedianFilter(7))
# Save image
blur_img.save("Saved_images/MedianFilter.jpg")
# Display image
blur_img.show()
```

## Median Filter using skimage

```python
# Import necessary libraries
from skimage import io
from skimage import filters
from skimage import data
from skimage.morphology import disk
```

```python
# Load image
img = io.imread("images/input.png")

# Apply filter 
out = filters.median(img,disk(7))

# Save and display image
io.imsave("Saved_images/MedianFilter scikit.jpg",out)
io.imshow(out)
io.show()
```

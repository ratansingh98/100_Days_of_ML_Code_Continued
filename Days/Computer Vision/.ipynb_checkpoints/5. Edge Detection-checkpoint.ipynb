## Using Sobel  Edge Detection method in scikit

```python
# Import neccessary libraries
from skimage import io, filters, color
```

```python
# Read image
img = io.imread("images/image.jpg")

# Convert RGB to gray
img = color.rgb2gray(img)

# For edge detection
edge = filters.sobel(img)

# Display and save image
io.imshow(edge)
io.imsave("Saved_images/sobel_edge.jpg", edge)
io.show()
```

## Using Canny Edge Detector method in scikit

```python
# Import neccessary libraries
from skimage import io, feature, color
```

```python
# Read image
img = io.imread("images/image.jpg")

# Convert RGB to gray
img = color.rgb2gray(img)

# For edge detection
edge = feature.canny(img,3)

# Display and save image
io.imshow(edge)
io.imsave("Saved_images/canny_edge.jpg", edge)
io.show()
```

## Using Hough Line Edge Detector method in scikit

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, draw, data
from skimage.feature import canny
from skimage.transform import hough_line,probabilistic_hough_line
from skimage.color import rgb2gray
```

```python
# Load image from data
image = data.camera()

# Edge detection
edges = canny(image, 2, 1, 25)
lines = hough_line(image)

# Check probablistic hough lines
probabilistic_lines = probabilistic_hough_line(edges, threshold= 10, line_length=5, line_gap=3)
probabilistic_lines
```

## Harris Corner Detection

```python
# Import necessary libararies
from matplotlib import pyplot as plt
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.feature import corner_harris, corner_subpix, corner_peaks
```

```python
# Load image
image = imread('images/image.png')
image = rgb2gray(image)

# Apply corner detection algorithm
corners = corner_harris(image)
coords = corner_peaks(corners, min_distance=5)
coords_subpix = corner_subpix(image, coords, window_size=13)
```

```python
# Diplay the image
fig, ax = plt.subplots()
ax.imshow(image, interpolation='nearest', cmap=plt.cm.gray)
ax.plot(coords[:,1], coords[:,0],'.b', markersize=3)
ax.plot(coords_subpix[:,1], coords_subpix[:,0], '+r', markersize=15)
ax.axis((0,350,350,0))
plt.savefig("Saved_images/Harris Detection.jpg")
plt.show()
```

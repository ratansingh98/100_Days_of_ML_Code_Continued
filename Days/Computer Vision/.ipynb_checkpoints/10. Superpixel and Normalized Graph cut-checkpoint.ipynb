## Superpixel

```python
# Import Necessary libraries
from skimage import segmentation, color
from skimage.io import imread
from skimage.future import graph
from matplotlib import pyplot as plt
```

```python
# Read image
img = imread('images/original.png')

# Apply superpixel operations
img_segments = segmentation.slic(img, compactness=20, n_segments=500)
superpixels = color.label2rgb(img_segments, img, kind='avg')
```

```python
# Display and store output
fig, ax = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True, figsize=(6, 8))

ax[0].imshow(img)
ax[1].imshow(superpixels)

for a in ax:
    a.axis('off')

plt.tight_layout()
plt.savefig("Saved_images/Superpixel.jpg")
plt.show()
```

## Normalize Graph cut

```python
# Import necessary libraries
from skimage import data, segmentation, color
from skimage.io import imread
from skimage import data
from skimage.future import graph
from matplotlib import pyplot as plt
```

```python
# Load the image
img = data.astronaut()

# Perform K-mean clustering
img_segments = segmentation.slic(img, compactness=30, n_segments=200)
out1 = color.label2rgb(img_segments, img, kind='avg')

# Create graph using that clusters
segment_graph = graph.rag_mean_color(img, img_segments, mode='similarity')
img_cuts = graph.cut_normalized(img_segments, segment_graph)
normalized_cut_segments = color.label2rgb(img_cuts, img, kind='avg')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True, figsize=(6, 8))

ax[0].imshow(img)
ax[1].imshow(normalized_cut_segments)

for a in ax:
    a.axis('off')

plt.tight_layout()
plt.savefig("Saved_images/Normalize Graph cut.jpg")
plt.show()
```

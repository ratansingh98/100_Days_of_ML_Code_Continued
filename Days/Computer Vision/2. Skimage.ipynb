## RGB to Gray

```python
# Import library
from skimage import color
# read image
img = io.imread("images/image.jpg")

# convert to gray scale image
gray = color.rgb2gray(img)

# Display image
io.imshow(gray)
io.show()
```

## RGB to HSV

```python
# Convert rgb to hsv
img_hsv = color.rgb2hsv(img)

# Display image
io.imshow(img_hsv)
io.show()
```

##  Draw Module

```python
# Necessary libraties
from skimage import draw
import numpy as np
```

#### For circle

```python
# Set the size of image
img = np.zeros((100,100),dtype=np.uint8)
# Size of circle
x, y = draw.circle(50, 50, 10)
img[x,y] = 1 

# Display circle
io.imshow(img)
io.show()
```

#### For ellipse

```python
# Size of ellipse
x, y = draw.ellipse(50, 50, 10, 20)
img[x,y] = 1 

# Display ellipse
io.imshow(img)
io.show()
```

#### For polygon

```python
r = np.array([10, 25, 80, 50])
c = np.array([10, 60, 40, 10])

# Size of polygon
x, y = draw.polygon(r, c)
img[x,y] = 1 

# Display polygon
io.imshow(img)
io.show()
```

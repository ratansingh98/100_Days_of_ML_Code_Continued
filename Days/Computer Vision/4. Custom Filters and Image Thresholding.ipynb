# Custom filter using kernal

```python
from PIL import Image
from PIL import ImageFilter
```

```python
# Load image
img = Image.open("images/image.jpg")

# Convert image into gray scale
img = img.convert("L")

# Apply custom filter
new_img = img.filter(ImageFilter.Kernel((3,3),[1,0,-1,5,0,-5,1,0,1]))

# Save and display image
new_img.save("Saved_images/create_filter_image.jpg")
img.show()
new_img.show()
```

## Thesholding

```python
# Import necessary libraries
import matplotlib.pyplot as plt
from skimage import data
from skimage.filters import threshold_otsu, threshold_local
from skimage.io import imread
from skimage.color import rgb2gray
```

```python
# Load and convert image to gray
image = imread("images/image.jpg")
image = rgb2gray(image)
```

```python
# Define threshold
global_thesh = threshold_otsu(image)
binary_global = image > global_thesh
block_size = 35
binary_adaptive = threshold_local(image, block_size, offset=10)
```

```python
# Create axes of subplots
fig, axes = plt.subplots(nrows=3, figsize=(7,8))
ax0, ax1, ax2 = axes
plt.gray()

# Plot the true image
ax0.imshow(image)
ax0.set_title("Image")

# Plot the binal global image
ax1.imshow(binary_global)
ax1.set_title("Binary Global")

# Plot the binal global image
ax2.imshow(binary_adaptive)
ax2.set_title("Binary Adaptive ")

for ax in axes:
    ax.axis('off')
plt.savefig("Saved_images/Threshold.jpg")
plt.show()
```
